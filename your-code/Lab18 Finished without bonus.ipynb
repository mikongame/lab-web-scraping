####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html,"lxml")
table = soup.find("tbody", {"id": "tbody"})
````

````python
rows = table.find_all("tr")
rows
````

````python
earthquakes = []
    
for earthquake in rows:
    date, time = earthquake.find("td", {"class": "tabev6"}).find("a").text.split()
    latitude_degree, longitude_degree = earthquake.find_all("td", {"class": "tabev1"})
    latitude_cardinality, longitude_cardinality, magnitude = earthquake.find_all("td", {"class": "tabev2"})
    latitude = f"{latitude_degree.text.strip()} {latitude_cardinality.text.strip()}"
    longitude = f"{longitude_degree.text.strip()} {longitude_cardinality.text.strip()}"
    region = earthquake.find('td', {'class': 'tb_region'}).text.strip()
    results = [date, time, latitude, longitude, region]
    earthquakes.append(results)
````

````python
colnames = ["Date", "Time", "Latitude", "Longitude", "Region"]

datos = earthquakes

earthquakeDF = pd.DataFrame(datos[:20], columns = colnames)

earthquakeDF
````
